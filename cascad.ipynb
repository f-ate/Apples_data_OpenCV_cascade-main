```python
import cv2
import numpy as np
cv2.namedWindow("Face Recognizer")
vc = cv2.VideoCapture(1)
   
font = cv2.FONT_HERSHEY_SIMPLEX
face_cascade = cv2.CascadeClassifier('cascade.xml')

while vc.isOpened():
        _, frame = vc.read()
        img = frame
        height, width, channels = frame.shape

        
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)

        # Loop through all the faces detected 
        identities = []
        for (x, y, w, h) in faces:
            x1 = x
            y1 = y
            x2 = x+w
            y2 = y+h
            face_image = frame[max(0, y1):min(height, y2), max(0, x1):min(width, x2)]
            
            
            imgResized = cv2.resize(face_image,(224,224))
            imgResized = np.asarray(imgResized)
            imgResized = imgResized.reshape((-1, 224, 224, 3))
            imgResized = imgResized.reshape(1,3,224,224).transpose(0,2,3,1)
            
            imgResized = imgResized / 255.0
            
            identity = np.argmax(model.predict(imgResized))
            
            
            if identity is not None:
                img = cv2.rectangle(frame,(x1, y1),(x2, y2),(255,255,255),2)
                cv2.putText(img, str(classes[identity.item()]), (x1+5,y1-5), font, 1, (255,0,0), 4)
                
            
        #print(identity)  
        #print(imgResized)
            
        key = cv2.waitKey(100)
        cv2.imshow("Face Recognizer", cv2.resize(img, (800, 600)))

        if key == 27: # exit on ESC
            break
vc.release()
cv2.destroyAllWindows()
```

```python
# working


import cv2
import numpy as np
cv2.namedWindow("Face Recognizer")
vc = cv2.VideoCapture(1)
   
font = cv2.FONT_HERSHEY_SIMPLEX
face_cascade = cv2.CascadeClassifier('new_data_cascade.xml')

while vc.isOpened():
        _, frame = vc.read()
        img = frame
        height, width, channels = frame.shape

        
        
        gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        faces = face_cascade.detectMultiScale(gray,
                                             scaleFactor = 5.5,
                                             minNeighbors = 91,
                                             minSize=(70,78))

        # Loop through all the faces detected 
        identities = []
        for (x, y, w, h) in faces:
            x1 = x
            y1 = y
            x2 = x+w
            y2 = y+h
            face_image = frame[max(0, y1):min(height, y2), max(0, x1):min(width, x2)]
            
            
            imgResized = cv2.resize(face_image,(224,224))
            imgResized = np.asarray(imgResized)
            imgResized = imgResized.reshape((-1, 224, 224, 3))
            imgResized = imgResized.reshape(1,3,224,224).transpose(0,2,3,1)
            
            imgResized = imgResized / 255.0
            classes = ['Fresh','Rotten']
            identity = model.predict(imgResized)
            print(int(identity.round()[0][0]))
            
            if identity is not None:
                img = cv2.rectangle(frame,(x1, y1),(x2, y2),(255,255,255),2)
                cv2.putText(img, str(classes[int(identity.round()[0][0])]), (x1+5,y1-5), font, 1, (0,200,0), 4)
                
            
        #print(identity)  
        #print(imgResized)
            
        key = cv2.waitKey(100)
        cv2.imshow("Face Recognizer", cv2.resize(img, (800, 600)))

        if key == 27: # exit on ESC
            break
vc.release()
cv2.destroyAllWindows()
```
